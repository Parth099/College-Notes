## 1. Example

### 1.1 Load data with pandas

```python
import numpy as np
import pandas as pd

# load data with pandas
df = pd.read_csv('cars.csv')
# print(df)

# convert a pandas dataframe to a numpy array
X = df.values[:, 1:-1]
print(type(X))
print(X.shape)
print(X)
```

### 1.2 Operation for vector

```python
#Get first row and the second to last row

x1 = X[0, :]
x2 = X[-2, :]

print(x1)
print(x2)

# Addition

y = x1 + x2
print(y)
```

### 1.3 Operation for matrix

```python
# Get first 5 rows and last 5 rows 

m1 = X[0:5, :]
m2 = X[-5:, :]

print(m1)
print(m2)

# Subtraction
m = m1 - m2
print(m)
```

## 2 Task

### 2.1 Get the first column and the second to last column from X

```python
# this is your code to get two vectors
v1 = X[:,  0]
v2 = X[:, -1]

print("Vector type: %s\n"%(type(v1)))
print(v1, v2, sep="\n\n")
```

### 2.2 Perform following operations on these two vectors and print the result

```python
from numpy import linalg as LA
import pandas as pd

# 1. addition and subtraction
print("add", v1 + v2, "\n")
# 2. element-wise multiplication
print("element-wise mult", np.multiply(v1, v2), "\n")
# 3. inner product
print("dot product", np.dot(v1, v2), "\n")
# 4. three norms
l1_norm = LA.norm(v1, ord=1) # sum of abs values
l2_norm = LA.norm(v1, ord=2) # Euclidean norm
inf_norm = LA.norm(v1, ord=np.inf) # largest element

norms = {
    "L1-norm": [l1_norm],
    "L2-norm": [l2_norm],
    "INF-norm": [inf_norm],
}

norms_frame = pd.DataFrame(data=norms)
print(norms_frame, "\n")

# 5. compute the distance between these two vectors
diff = v1 - v2
dist = LA.norm(diff, ord=2)
print("Euclidean Distance between v1 and v2: ", dist)
```
