# Lab 2: Data Preprocessing

In this assignment, we will learn how to explore the raw data and preprocess it. 

The dataset we are going to exlore is an insurance data. It provides different features of each user as follows:
* age: age of the user
* sex: gender of the user
* bmi: body mass index, providing an understanding of body
* children: number of children covered by health insurance / number of dependents
* smoker: smoker or not
* region: the user's residential area in the US, northeast, southeast, southwest, northwest.

Additionally, the medical cost of each user is also provided: 
* charges: the medical cost 

Please follow Lecture 5_data_understanding and Lecture 6_data_preprocessing to complete following questions. 

### Q1. Load data with Pandas and output the basic information of this dataset, such as the features and their data types.  Which features are numerical features and which users are categorical features?


```python
import pandas as pd
import numpy as np

# load dataset
df = pd.read_csv("insurance.csv")

numericals = df._get_numeric_data().columns;
categoricals = [i for i in df.columns if df.dtypes[i] == 'O']

df.info()
```

### Q2. Check whether there are missing values in this dataset.

### Q4. Use corr() function of Pandas to show the correlation between different numerical features

```python
# util corr over numerical features only
hcorr = df.corr(numeric_only=True)
hcorr.style.background_gradient()
```

### Q5. For all categorical features, use bar plot to visualize the number of users within each category.

```python
# Create just a figure and only one subplot
fig, ax = plt.subplots(1, 3, sharey=False, sharex=False)
fig.tight_layout()

# plot 3 via loops
for i ,(plot, label) in enumerate(zip(ax, categoricals)):
    # generate bar charts using ranges and values
    plot.bar(np.unique(df[label].values), df[label].value_counts(), align='center')
    
    # start working on current axis
    plt.sca(ax[i])
    
    # rotate ticks to improve readablity
    plt.xticks(rotation=45)
```

### Q6. Convert all categorical features into numerical features with Label Encoding or One-Hot Encoding

```python
# 1. sex
# 2. smoker
# 3. region

# One-hot Encoding
from sklearn import preprocessing

Encoder = preprocessing.OneHotEncoder(sparse_output=False)

for label in categoricals:
    # fit to data to get transformed copy and new feature names
    encoded_feature = Encoder.fit(df[[label]])
    transformed_feature = Encoder.transform(df[[label]])
    new_column_names = encoded_feature.get_feature_names_out()

    # generate new df 
    ohe_encoded = pd.DataFrame(transformed_feature, columns=new_column_names)

    # drop old column
    df = pd.concat([df, ohe_encoded], axis=1).drop([label], axis=1)

# debug printout
# df
```

### Q7. Normalize all numerical features

```python
# now all features are numercial due to prev cell

# numerical features
# 1. age
# 2. bmi
# 3. Children
# 4. charges

for label in numericals:
    # normalize process (z score)
    vector = df[label].values.astype(dtype=np.float32)
    mean = np.mean(vector, axis=0)
    std = np.std(vector, axis=0)

    vector = (vector - mean) / std

    # drop existing col
    df.drop([label], axis=1)

    #add new normalized col
    df[label] = vector
```

### Q8. Save your preprocessed data into a csv file. Submit your code and the preprocessed data.

```python
# index = write row names
df.to_csv("./preprocessed-ohe-z_score-insurance.csv",index=False)
```
